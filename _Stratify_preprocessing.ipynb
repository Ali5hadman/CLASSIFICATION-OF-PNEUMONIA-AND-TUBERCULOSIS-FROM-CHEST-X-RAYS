```python
from PIL import Image
import scipy


def processing(path,imagename):
    image = Image.open(path+'/'+imagename)
    data = np.asarray(image)
    datafiltered=scipy.ndimage.median_filter(data,(3,3))
    datafiltered = scipy.ndimage.gaussian_filter(datafiltered, sigma=1)
    image2 = Image.fromarray(datafiltered)
    image2=image2.resize((200,200))
    image2.save('train2/'+imagename)
```

```python
for file in os.listdir("train"):
    processing('train',file)
```
